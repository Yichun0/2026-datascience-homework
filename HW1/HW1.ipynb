## Problem 1: Percent Negative

Write a function that returns what percentage of a list is negative.

You can assume the list contains only numbers and is not empty.

```python
# your code
def percent_negative(list):
    if len(list) == 0:
        return 0
    total_count = len(list)
    neg_count = 0
    for item in list:
        if item < 0:
            neg_count += 1
    return (neg_count/total_count)*100

```

## Problem 2: Reversing & Removing BAD_VALUE in a List (Two Ways)

Write two functions that returns the reverse of a given list with all instances of "BAD_VALUE" removed.

The first function should use element-wise processing by list index. The second should use slicing within a list comprehension.

```python
# your code
def fixreverse_elem(list, keyword = "BAD_VALUE"):
    if len(list) == 0:
        return list
    reverse_list = []
    for idx in range(len(list)-1, -1,-1):
        if list[idx] != keyword:
            reverse_list.append(list[idx])
    return reverse_list


def fixreverse_slc(list,keyword = "BAD_VALUE"):
    if len(list) == 0:
        return list
    cleaned_list = [item for item in list if item != keyword]
    return cleaned_list[::-1]

```

## Problem 3: Combine List Pairwise Lexicographically

Write a function that combines two lists by pairwise taking the elements of each list, determining which comes first based on lexicographical order according to Python, e.g., ['a','b','c',4], [1,2,3,'d'] → [1,'a',2,'b',3,'c',4,'d'].

In case of two lists of different length, the elements of the longer list are appended at the end, e.g., [1,2,3,4],[10,20]=>[1,10,2,20,3,4].

```python
# your code
def combine_list(list1, list2):
    ordered_list1 = sorted(list1,key=str)
    ordered_list2 = sorted(list2,key=str)
    len_1 = len(ordered_list1)
    len_2 = len(ordered_list2)
    if len_1 == 0:
        return ordered_list2
    elif len_2 == 0:
        return ordered_list1
    else:
        combo_list = []
        if len(ordered_list1) == len(ordered_list2): 
            for item1, item2 in zip(ordered_list1,ordered_list2):
                if type(item1) == "int":
                    combo_list.append(item1)
                    combo_list.append(item2)
                
        else:
            shorter_len = min(len_1, len_2)
            if len_1 != shorter_len:
                longer_list = ordered_list1
            else:
                longer_list = ordered_list2
            for idx in range(shorter_len):
                combo_list.append(ordered_list1[idx])
                combo_list.append(ordered_list2[idx])
            for item in longer_list[-abs(len_1-len_2):]:
                combo_list.append(item)
        return combo_list
            

```

```python
# call your function with a couple of different lists and print the results here
print(combine_list([1, 1, 2, 30, 5, 8], ['w', 'x', 'y', 3]))
print(combine_list(['delta', 'echo', 'foxtrot', 'golf'], [2, 7, 1, 8, 2, 8]))
```

## Problem 4: Largest Digit

Write a function that returns the largest digit in a number, e.g., 783456 → 8. Assume the number is round (has no decimal places).

```python
# your code
def largest_digit(num):
    if num < 0:
        str_num = str(num)[1:]
    else:
        str_num = str(num)
    largest = 0
    for item in str_num:

        if int(item) > largest:
            largest = int(item)
    return largest
```

## Problem 5: Check Sorting in Three Ways

Write three functions that check if a list is sorted, returning True or False. The first function should use a for-loop, the second a while-loop, and the third should use recursion.

*Hint: A function may need more than one argument, but you should allow the function to be called with only one argument given.*

```python
# your code
def check_for(list):
    if len(list) == 0:
        return "Empty list"
    previous = list[0]
    for item in list[1:]:
        if item < previous:
            return False
    return True

def check_while(list):
    if len(list) == 0:
        return "Empty list"
    previous = list[0]
    idx = 1
    status = False
    while idx <= len(list) -1:
        if list[idx] < previous:
            status = False
            break
        else:
            idx += 1
            status = True
    return status

def check_recursive(list):
    if len(list) == 0:
        return "Empty list"
    if len(list) == 1:
        return True
    previous = list[0]
    if previous > list[1]:
        return False
    return check_recursive(list[1:])
```

```python
# call your function with a couple of different lists and print the results here
list1 = range(1,9)
list2 = [1, 6, 2, -2, -2, 5, 9, 3, 7, -4, 9]
list3 = [-8, -2, 0, 1, 1, 2, 8]
list4 = [3, 6, 9, -3, 2, 7, 6, 1, -9]

print(check_for(list1), check_while(list1), check_recursive(list1))
print()
print(check_for(list2), check_while(list2), check_recursive(list2))
print()
print(check_for(list3), check_while(list3), check_recursive(list3))
print()
print(check_for(list4), check_while(list4), check_recursive(list4))
```

## Problem 6: Filtering

Write a function `filter_to_numbers` that filters out elements from a list that are not numbers. Preserve the order of the list.

*Hint: Look up the `isinstance` function*

```python
# your code
def filter_to_numbers(list):
    if len(list) == 0:
        return []
    for item in list[:]:
        if not isinstance(item,int):
            list.remove(item)
    return list
```

## Problem 7: Hofstadter Q-Sequence

Write a function that takes a number `n` and an empty array and then writes the first `n` [Hofstadter Q-Sequence](https://en.wikipedia.org/wiki/Hofstadter_sequence#Hofstadter_Q_sequence) into the empty array. Print the array. 

The sequence can be generated via a relation where $Q(n) = Q(n - Q(n-1)) + Q(n - Q(n-2))$ with $Q(1) = Q(2) = 1$. Note this sequence starts with $Q(1)$.

Thus, if you call `hofstadterq(5)`, it should thus produce `[1, 1, 2, 3, 3]`.

```python
# your code  
def hofstadterq(n, seq):
    if n < 0:
        return []
    seq = [1,1] 
    if n < 3:
        return seq
    else:
        for i in range(3,n+1): # index of the sequence not the python index
            # start to calculate new values from Q(3)
            value = seq[i-1 - seq[i-2]] + seq[i-1- seq[i-3]] # reduce one more for python indexing 
            seq.append(value)
        return seq

    

```

# Problem 8: Screentime Moving Maximum

Write a function `max_k_window` that takes a list of hours spent viewing a screen every day and computes a moving maximum over every consecutive `k` days. Some days will be missing measurements and will instead have an empty string for a value.

**Caveats:**
* The days missing measurements should be included in your `k` day window, but should not affect maximum.
  * e.g.`max_k_window([8.7,'',9.2],3)` should return `[9.2]`
* If all days in a window are missing measurements then the maximum for that window should also have the value `None`.
  * e.g. `max_k_window(['', '', 5.3], 2)` should return `[None, 53]`
* Only compute the maximum as long as you have enough data points to fill the window.
  * e.g. `max_k_window([10.0,2.5,3.0,4.5,5.0,6.0,7.5,8.0],7)` should return a list of length 2.
  * e.g. `max_k_window([12.0,24.0,3.0,4.5,5.5],2)` should return a list of length 4.
  * e.g. `max_k_window([10.0,20.0,3.5], 4)` should return an empty list.

You can use these as example lists, but make sure that your algorithm works on other lists/windows.

```python
# your code 
def max_k_window(time_ls,k):
    max_time = []
    starting_idx = 0
    end_idx = k
    max = 0
    if k > len(time_ls):
        return []
    while end_idx < len(time_ls) + k:
        if end_idx > len(time_ls): # end of the list
            end_idx == len(time_ls) - 1
        window = time_ls[starting_idx: end_idx]
        for time in window:
            if time == "":
                continue
            if time > max:
                max = time
        if max == 0:
            max_time.append(None)
        else:
            max_time.append(max)
        max = 0
        starting_idx += k
        end_idx += k
    return max_time
    
    
        
```
